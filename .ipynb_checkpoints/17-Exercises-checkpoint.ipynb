### Exercise 1. 

Read Total Profit for all months and show it using a line plot with Month Number as your baseline

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
import pandas as pd
import matplotlib.pyplot as plt

define=pd.read_csv("./data/company_sales_data.csv") #how you read a csv nested in juypter
profit= define ['total_profit'].tolist() #data is cast to a list
month = define ['month_number'].tolist()
plt.plot(month,profit, label="Month Profit of last year") #parameters include the two lists that were cast
plt.xlabel('Months')
plt.ylabel('Profit $')
plt.xticks(month)
plt.title('Profit Per Month')
plt.yticks([100000, 200000,300000, 400000, 500000])
plt.show()
````

````python


plt.plot(month,profit, label="Month Proit of last year", color='r', marker='o', markerfacecolor='k',linestyle='--', linewidth=3) 
plt.show()
#parameters include the two lists that were cast
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
#to include the new products on the plot you must cast them to a list as well
faceCream= define ['facecream'].tolist()
faceWash = define ['facewash'].tolist()
toothPaste = define ['toothpaste'].tolist()
bathingsoap = define ['bathingsoap'].tolist()
shampoo= define ['shampoo'].tolist()
moisturizer = define ['moisturizer'].tolist()
#check spelling of feilds in data set and make sure they match with what youre calling youre list
plt.plot(month, faceCream, label='Face cream', marker='o', linewidth=3)
plt.plot(month, faceWash, label='Face wash', marker='o', linewidth=3)
plt.plot(month, toothPaste, label='toothPaste', marker='o', linewidth=3)
plt.plot(month, bathingsoap, label='bathingsoap', marker='o', linewidth=3)
plt.plot(month, shampoo, label='shampoo', marker='o', linewidth=3)
plt.plot(month, moisturizer, label='moisterizer', marker='o', linewidth=3)
plt.legend(loc='upper left')
plt.show()
         
````

### Exercise 4.

Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should “–“.

````python
x=month
y=toothPaste
plt.scatter(x, y, label='ToothPaste')
plt.xlabel("Month")
plt.ylabel("units sold")
plt.title('ToothPaste Sales')
plt.xticks(month)
plt.grid(True, linewidth=1, linestyle="--")
plt.show()
````

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
x=month
y=faceWash
z=faceCream
plt.bar([a-0.25 for a in x],y,width=0.25, label="FaceWash sales", align="edge")
plt.bar([a+0.25 for a in x],z,width=-0.25, label="FaceCream sales", align="edge")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.title("FaceWash Sales YTD")
plt.xticks(month)
plt.ylabel('Sales units in number')

````

### Exercise 6.
Write a script that creates your favorite visual type you have learned about, and display the visual once from within the notebook:

````python
import numpy as np
from matplotlib import pyplot as plt

# Creating equally spaced 100 data in range 0 to 2*pi
theta = np.linspace(0, 2 * np.pi, 100)

# Generating x and y data
x = 16 * ( np.sin(theta) ** 3 )
y = 13 * np.cos(theta) - 5* np.cos(2*theta) - 2 * np.cos(3*theta) - np.cos(4*theta)

# Plotting
heart=plt.plot(x, y)
plt.savefig('heart')
plt.title('Heart Shape')
plt.show()
````
